```python
textToSearch = 'ahir bhairav'
textToSearch  = textToSearch.replace(" ","+")
print(textToSearch)
```

```python
response = requests.get("https://www.youtube.com/results?search_query={}&sp=EgIQAQ%253D%253D".format(textToSearch)).text
```

```python
soup = BeautifulSoup(response,'lxml')
# print(soup)
```

```python
script = soup.find_all("script")[32]
json_text = re.search('var ytInitialData = (.+)[,;]{1}',str(script)).group(1)
json_data = json.loads(json_text)
content = json_data['contents']['twoColumnSearchResultsRenderer']['primaryContents']['sectionListRenderer']['contents'][0]['itemSectionRenderer']['contents']

# with open("script.txt","w",encoding="utf-8") as output:
#    output.write(json_text)
```

```python
video_metadata = []
for video in content:
    title = video['videoRenderer']['title']['runs'][0]['text']
    published = video['videoRenderer']['publishedTimeText']['simpleText']
    duration =  video['videoRenderer']['lengthText']['simpleText']
    url = video['videoRenderer']['navigationEndpoint']['commandMetadata']['webCommandMetadata']['url']
    video_metadata.append({'title' :title, 'published' :published, 'duration':duration, 'url':url})
```
